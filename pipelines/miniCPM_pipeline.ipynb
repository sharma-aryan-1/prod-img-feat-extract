```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import nltk
nltk.download('averaged_perceptron_tagger_eng') # run first time
nltk.download('punkt_tab') # run first time

# POS tag and filter
def pos_extract(text):
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)

    for i in range(len(tagged) - 1):
      word, tag = tagged[i]
      if tag == "CD":
          return word, tagged[i + 1][0]

text = ""
for word in pos_extract("The bat is 4 kgs"):
  text += word

if text[-1] == 's' or 'S':
  text = text[:-1]

print(text)


# conver abbr to full form
import re
import pint

ureg = pint.UnitRegistry()

def expand_units_in_sentence(sentence):
    pattern = r'(\d+(?:\.\d+)?)[ ]*([A-Za-z]+)'
    def repl(match):
        number_str, unit_abbr = match.groups()
        try:
            number = float(number_str)
            quantity = number * ureg(unit_abbr)
            return str(quantity)
        except Exception:
            return match.group(0)

    return re.sub(pattern, repl, sentence)

expanded_text = expand_units_in_sentence(text)
print(expanded_text)
```

```python
def numtofloat(text):
    # Use regex to find all numbers in the string
    numbers = re.findall(r'\d+\.?\d*', text)

    # Replace each number with its float representation
    for number in numbers:
        text = text.replace(number, str(float(number)))

    return text


import editdistance

def charErrorRate(reference, hypothesis):
    edit_distance = editdistance.eval(reference, hypothesis)
    return edit_distance / len(reference) if len(reference) > 0 else 0.0
```

```python
def normalize_response(response, entity_name, entity_unit_map, conversion_dict):
    # Handle ranges (e.g., 175-265V)
    range_match = re.search(r'(\d+\.?\d*)\s*[-–—]\s*(\d+\.?\d*)', response)
    if range_match:
        val1, val2 = range_match.groups()
        unit = extract_unit(response, entity_name, conversion_dict)
        return f"[{float(val1):.2f}, {float(val2):.2f}] {unit}"

    # Extract exact numeric value
    value_match = re.search(r'(\d+\.?\d*)', response.replace(',', ''))
    value_str = value_match.group(1) if value_match else None

    # Normalize units with priority
    normalized_response = response.lower()
    ordered_patterns = sorted(conversion_dict.keys(), key=lambda x: (-len(x), x))
    for pattern in ordered_patterns:
        normalized_response = re.sub(pattern, conversion_dict[pattern], normalized_response)

    # Extract valid units
    valid_units = entity_unit_map[entity_name]
    found_units = [unit for unit in valid_units if re.search(rf'\b{re.escape(unit)}\b', normalized_response)]

    # Format output
    if value_str:
        # Preserve original decimal precision
        formatted_value = f"{float(value_str):.2f}".rstrip('0').rstrip('.') if '.' in value_str else f"{value_str}.0"
        unit = found_units[0] if found_units else valid_units[0]
        return f"{formatted_value} {unit}"

    return f"0.0 {valid_units[0]}"


def extract_unit(text, entity_name, conversion_dict):
    # Helper function for range handling
    temp_text = text.lower()
    for pattern in conversion_dict:
        if re.search(pattern, temp_text):
            return conversion_dict[pattern].strip()
    return entity_unit_map[entity_name][0]
```
