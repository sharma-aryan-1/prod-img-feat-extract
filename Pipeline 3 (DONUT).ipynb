```python
# @title Regex
import re
entity_unit_map = {
    'width': ['centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'],
    'depth': ['centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'],
    'height': ['centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'],
    'item_weight': ['gram', 'kilogram', 'ounce', 'pound', 'microgram', 'milligram', 'ton'],
    'maximum_weight_recommendation': ['gram', 'kilogram', 'ounce', 'pound', 'microgram', 'milligram', 'ton'],
    'voltage': ['volt', 'kilovolt', 'millivolt'],
    'wattage': ['watt', 'kilowatt'],
    'item_volume': ['litre', 'millilitre', 'centilitre', 'decilitre', 'microlitre',
                    'gallon', 'fluid ounce', 'pint', 'quart', 'cup', 'cubic foot', 'cubic inch']
}
allowed_units = {unit for entity in entity_unit_map for unit in entity_unit_map[entity]}

conversion_dict = {
    # ---------------------------------------------------------------
    # Multi-character units (longest first to prevent partial matches)
    # ---------------------------------------------------------------
    # Length/Distance
    r'(?<=\d)cubic in\b': ' cubic inch',
    r'(?<=\d)cubic ft\b': ' cubic foot',
    r'(?<=\d)millimetre\b': ' millimetre',  # Explicit full-word matches
    r'(?<=\d)centimetre\b': ' centimetre',
    r'(?<=\d)imp gal\b': ' imperial gallon',
    r'(?<=\d)fl oz\b': ' fluid ounce',
    r'(?<=\d)mm\b': ' millimetre',
    r'(?<=\d)cm\b': ' centimetre',
    r'(?<=\d)ft\b': ' foot',
    r'(?<=\d)yd\b': ' yard',
    r'(?<=\d)in\b': ' inch',
    r'(?<=\d)inch\b': ' inch',
    r'(?<=\d)"': ' inch',
    r'(?i)(?<=\d)kgs?\b': ' kilogram',  # Handles KG/Kgs/kgs
    r'(?i)(?<=\d)lbs?\b': ' pound',     # Handles LB/Lbs/lbs
    r'(?<=\d)"': ' inch',               # Handles inches symbol
    r'(?i)inches\b': ' inch',
    # Weight/Mass
    r'(?<=\d)kilogram\b': ' kilogram',
    r'(?<=\d)milligram\b': ' milligram',
    r'(?<=\d)microgram\b': ' microgram',
    r'(?<=\d)kg\b': ' kilogram',
    r'(?<=\d)mg\b': ' milligram',
    r'(?<=\d)µg\b': ' microgram',
    r'(?<=\d)oz\b': ' ounce',
    r'(?<=\d)lb\b': ' pound',

    # Voltage
    r'(?<=\d)([vV])\b': ' volt',
    r'(?<=\d)kilovolt\b': ' kilovolt',
    r'(?<=\d)millivolt\b': ' millivolt',
    r'(?<=\d)kV\b': ' kilovolt',
    r'(?<=\d)mV\b': ' millivolt',
    r'(?<=\d)V\b': ' volt',


    # Wattage
    r'(?<=\d)kilowatt\b': ' kilowatt',
    r'(?<=\d)kW\b': ' kilowatt',
    r'(?<=\d)W\b': ' watt',

    # Volume
    r'(?<=\d)microlitre\b': ' microlitre',
    r'(?<=\d)millilitre\b': ' millilitre',
    r'(?<=\d)centilitre\b': ' centilitre',
    r'(?<=\d)decilitre\b': ' decilitre',
    r'(?<=\d)µl\b': ' microlitre',
    r'(?<=\d)ml\b': ' millilitre',
    r'(?<=\d)cl\b': ' centilitre',
    r'(?<=\d)dl\b': ' decilitre',
    r'(?<=\d)gal\b': ' gallon',
    r'(?<=\d)pint\b': ' pint',
    r'(?<=\d)quart\b': ' quart',
    r'(?<=\d)cup\b': ' cup',

    # ---------------------------------------------------------------
    # Single-character units (handle last to prevent conflicts)
    # ---------------------------------------------------------------
    # Length/Distance
    r'(?<=\d)m\b': ' metre',  # Only matches when preceded by digit

    # Weight/Mass
    r'(?<=\d)g\b': ' gram',
    r'(?<=\d)ton\b': ' ton',

    # Volume
    r'(?<=\d)l\b': ' litre',
}

def numtofloat(text):
    # Use regex to find all numbers in the string
    numbers = re.findall(r'\d+\.?\d*', text)

    # Replace each number with its float representation
    for number in numbers:
        text = text.replace(number, str(float(number)))

    return text


import editdistance

def charErrorRate(reference, hypothesis):
    edit_distance = editdistance.eval(reference, hypothesis)
    return edit_distance / len(reference) if len(reference) > 0 else 0.0
```

```python
def normalize_response(response, entity_name, entity_unit_map, conversion_dict):
    # Handle ranges (e.g., 175-265V)
    range_match = re.search(r'(\d+\.?\d*)\s*[-–—]\s*(\d+\.?\d*)', response)
    if range_match:
        val1, val2 = range_match.groups()
        unit = extract_unit(response, entity_name, conversion_dict)
        return f"[{float(val1):.2f}, {float(val2):.2f}] {unit}"

    # Extract exact numeric value
    value_match = re.search(r'(\d+\.?\d*)', response.replace(',', ''))
    value_str = value_match.group(1) if value_match else None

    # Normalize units with priority
    normalized_response = response.lower()
    ordered_patterns = sorted(conversion_dict.keys(), key=lambda x: (-len(x), x))
    for pattern in ordered_patterns:
        normalized_response = re.sub(pattern, conversion_dict[pattern], normalized_response)

    # Extract valid units
    valid_units = entity_unit_map[entity_name]
    found_units = [unit for unit in valid_units if re.search(rf'\b{re.escape(unit)}\b', normalized_response)]

    # Format output
    if value_str:
        # Preserve original decimal precision
        formatted_value = f"{float(value_str):.2f}".rstrip('0').rstrip('.') if '.' in value_str else f"{value_str}.0"
        unit = found_units[0] if found_units else valid_units[0]
        return f"{formatted_value} {unit}"

    return f"0.0 {valid_units[0]}"
def extract_unit(text, entity_name, conversion_dict):
    # Helper function for range handling
    temp_text = text.lower()
    for pattern in conversion_dict:
        if re.search(pattern, temp_text):
            return conversion_dict[pattern].strip()
    return entity_unit_map[entity_name][0]
```

## Generate

Finally, we let the model autoregressively generate the answer to the question.

```python
import torch
import re

import time

start_time = time.time()

references=[]
predictions=[]
op = []
extracted_values = []
for i in range (NO_OF_SAMPLES):

  task_prompt = "<s_docvqa><s_question>{user_input}</s_question><s_answer>"
  pixel_values
  question = "what is the item_" + random_sample.loc[i, "entity_name"] + "?"
  prompt = task_prompt.replace("{user_input}", question)
  decoder_input_ids = processor.tokenizer(prompt, add_special_tokens=False, return_tensors="pt")["input_ids"]

  device = "cuda" if torch.cuda.is_available() else "cpu"
  model.to(device)

  outputs = model.generate(
    pixval[i].to(device),
    decoder_input_ids=processor.tokenizer(
        task_prompt,
        add_special_tokens=False,
        return_tensors="pt"
    ).input_ids.to(device),
    max_length=128,
    temperature=1.0,          # Controls randomness (0.1-1.5)
    top_p=0.95,               # Nucleus sampling threshold (0.7-0.95)
    do_sample=True,           # Enable sampling mode
    num_beams=1,              # Disable beam search
    pad_token_id=processor.tokenizer.pad_token_id,
    eos_token_id=processor.tokenizer.eos_token_id,
    use_cache=True,
    return_dict_in_generate=True,  # Explicitly request dictionary
    output_scores=True,
    output_attentions=True,  # Critical for attention extraction
  )
  # Decode generated output
  seq = processor.batch_decode(outputs.sequences)[0]

  # Remove special tokens
  seq = seq.replace(processor.tokenizer.eos_token, "").replace(processor.tokenizer.pad_token, "")

  # Extract answer using regex
  match = re.search(r"<s_answer>(.*?)</s_answer>", seq)
  answer = match.group(1).strip() if match else "N/A"

  extracted_values.append(answer)
  converted = normalize_response(answer, random_sample.loc[i, "entity_name"], entity_unit_map, conversion_dict)
  random_sample.loc[i, "converted"] = converted
  references.append(random_sample.loc[i, "entity_value"])
  predictions.append(converted)

# # Add extracted answers to the DataFrame
random_sample["predicted_answer"] = extracted_values
for i in range(5):
  random_sample.loc[i, "CER"] = charErrorRate(random_sample.loc[i, "entity_value"], random_sample.loc[i, "converted"])

# Print to verify
print(random_sample[["entity_name", "predicted_answer", "converted", "entity_value", "CER"]])

end_time = time.time()

execution_time = (end_time - start_time) / 50.0
print(f"Execution time: {execution_time:.4f} seconds")

```
